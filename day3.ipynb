```python
import os
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
import matplotlib

# Use an interactive backend
matplotlib.use('TkAgg')

# Paths
annoFolder = "/Users/niteshyadav/edge ai/labels"
imgFolder = "/Users/niteshyadav/edge ai/images"
polyLinedFolder = "/Users/niteshyadav/edge ai/multipolygon"

# Get the list of images
imgList = os.listdir(imgFolder)

# Process each image
for im in imgList:
    img = cv.imread(os.path.join(imgFolder, im))
    plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
    plt.title("Select Points: Click on Top-Left Corner to Quit")
    
    all_points = []  # List to store all polygons for the current image
    
    while True:
        points = []  # List to store points for one polygon
        plt.show(block=False)
        
        while True:
            # Select a single point (x, y)
            p = plt.ginput(1)
            if len(p) == 0:
                break
            
            px, py = int(p[0][0]), int(p[0][1])
            
            # Check if the user clicked on the top-left corner to quit
            if px < 50 and py < 50:
                plt.close()
                break
            
            # Append the point to the polygon list
            points.append((px, py))
            
            # Display the updated image with selected points
            plt.scatter(px, py, c='r')
            plt.draw()
        
        # If the user clicked on the top-left corner, exit the outer loop
        if len(points) == 0 or (px < 50 and py < 50):
            break
        
        # Append the completed polygon to the list of polygons
        all_points.append(points)
    
    plt.close()

    # Save the annotation (all polygons) to a text file
    dxt_txt = im.split(".")[0] + ".txt"
    with open(os.path.join(annoFolder, dxt_txt), 'a') as f:
        for poly in all_points:
            f.write(f"{poly}\n")
    
    # Draw the polygons on the image
    for poly in all_points:
        cv.polylines(img, [np.array(poly)], color=(0, 0, 255), isClosed=True, thickness=5)
    
    # Save the image with drawn polygons
    linedImg_txt = im
    cv.imwrite(os.path.join(polyLinedFolder, linedImg_txt), img)
    
    # Show the result
    cv.imshow("Image", img)
    cv.waitKey(0)
    cv.destroyAllWindows()
```
